```python
import pandas as pd
import numpy as np
import requests
import bs4 as bs
import urllib.request
import json
import tqdm

from webscrape_tools import *
```

```python
detail_list = get_links('https://www.homedepot.com/sitemap/d/pip_sitemap.xml')
detail_list[0:5]
```

```python
detail0 = get_links(detail_list[0])
detail0[0:5]
```

```python
#parsing product pages for detail0[10]
product = detail0[10]
source = urllib.request.urlopen(product).read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
product_description = soup.find('script', id='thd-helmet__script--productStructureData')
product_description
```

```python
description_json = json.loads(product_description.contents[0])
description_json
"""@type
name
description
depth
height
width
color
brand - name
aggregateRating - ratingValue, ratingCount
offers - url, price
"""
```

```python
data = description_json

reorganized_data = {
    '@type': data['@type'],
    'name': data['name'],
    'description': data['description'],
    'depth': data['depth'],
    'height': data['height'],
    'width': data['width'],
    'color': data['color'],
    'brandName': data['brand']['name'],
    'ratingValue': data['aggregateRating']['ratingValue'],
    'reviewCount': data['aggregateRating']['reviewCount'],
    'url': data['offers']['url'],
    'price': data['offers']['price'],
    'priceCurrency': data['offers']['priceCurrency']
}
```

```python
json_entries=[]

json_entries.append(reorganized_data)
```

```python
df = pd.DataFrame.from_records(json_entries)
```
